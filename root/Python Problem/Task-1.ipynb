```python
#importing python libraries
import tweepy
import jsonlines
import numpy as np
import pandas as pd
#getting tokens and keys
consumer_key = "XXXXXXXXXXXXXXX"
consumer_secret = "XXXXXXXXXXXXXXX"
access_token = "XXXXXXXXXXXXXXX"
access_token_secret = "XXXXXXXXXXXXXXX"
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
# Setting your access token and secret
auth.set_access_token(access_token, access_token_secret)
# Creating the API object while passing in auth information
api = tweepy.API(auth) 
```

```python
# The Twitter user who we want to get tweets from
name = "midasIIITD"
# Number of tweets to pull
tweetCount = 100

# Calling the user_timeline function with our parameters
results = api.user_timeline(id=name, count=tweetCount)
```

```python
with jsonlines.open("results.jsonl",mode='w') as writer:
    for tweet in results:
        writer.write(tweet._json)
        
#making  2-D array to store all the information 
answer=[]
with jsonlines.open("results.jsonl",mode='r') as reader:
    for item in reader.iter():
        temp=[]
        temp.append(item['text'])
        temp.append(item['created_at'][:16])
        temp.append(item['favorite_count'])
        temp.append(item['retweet_count'])
        temp.append(item['entities'])
        answer.append(temp)
        
#creating dataframe
df=pd.DataFrame(answer,columns=['text','data','likes','retweets','entities'])
```
